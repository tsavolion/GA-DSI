```python
import numpy as np
import scipy.stats as stats
import csv
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import random as random
```

```python
bb = pd.read_csv('/Users/supermaxfederman/DSI-NYC-2/projects/projects-weekly/project-02/assets/billboard.csv', encoding='latin-1')
```

```python
## Hypothesis: On average, the most popular songs of the year (the songs at the top of the chart the longest) ##
## will be released early in the Summer and peak in the Summer. ##

## Method: A chart showing how many songs are released and how many songs peak by the month will reveal the ##
## months with the top songs released. ##
```

```python
##acquire data
bb
```

```python
##Specify the data we need 
bb['date.entered'] = pd.to_datetime(bb['date.entered'])
```

```python
bb2= bb.groupby([bb['date.entered'].dt.year, bb['date.entered'].dt.month])['date.entered'].count().to_frame()
```

```python
bb2['date.entered'].plot(kind='bar')

## This chart shows the frequency of top song releases. It clearly shows that most top songs are released in
## the Winter to the Spring, with a few towards the end of Summer. 

### add title and axis titles/ labels
```

```python
##Specify the data we want
bb['date.peaked'] = pd.to_datetime(bb['date.peaked'])
```

```python
bbpeak = bb.groupby([bb['date.peaked'].dt.year, bb['date.peaked'].dt.month])['date.peaked'].count().to_frame()
```

```python
bbpeak['date.peaked'].plot(kind='bar')

## This chart shows the frequency of songs peaking in the charts by the month. It clearly illustrates that
## Winter is by far the most popular time of years for songs to peak in the charts. 

### add title/ subtitles
```

```python
bbpeak.sort_values('date.peaked')[-3:]
```

```python
## The months where the most songs peaked were January, December, and September in that order. 
```
