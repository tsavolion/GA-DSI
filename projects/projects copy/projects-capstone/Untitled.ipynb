```python
scaleddf = preprocessing.scale(X)
```

```python
vardf = cluster.KMeans(n_clusters=3)
vardf.fit(scaleddf)
labels = vardf.labels_
```

```python
for x in range(2,20): 
    vardf = cluster.KMeans(n_clusters=x)
    vardf.fit(scaleddf)
    labels = vardf.labels_
    print(x)
    print(metrics.silhouette_score(X, labels, metric='euclidean'))
    print(vardf.inertia_)
```
