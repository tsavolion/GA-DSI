```python
### 6
rows = []

with open(sat_scores_path, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)
header = rows[0]
print(header)
```

```python
### 4
data = rows[1:]
###5
print(data)
```

```python
sat_data = {}
for index, column_name in enumerate(header):
    sat_data[column_name]=[]
    for row in data:
        sat_data[column_name].append(row[index])
```

```python
states = sat_data['State'][0:52]
rates = sat_data['Rate'][0:52]
verbs = sat_data['Verbal'][0:52]
maths = sat_data['Math'][0:52]
```
