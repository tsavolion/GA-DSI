```python
def get_top_250():
    response = requests.get('http://www.imdb.com/chart/top')
    html = response.text
    entries = re.findall("<a href.*?/title/(.*?)/", html)
    return list(set(entries))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
%matplotlib inline
```

```python
from sklearn.preprocessing import MinMaxScaler

x = top_films4.loc[:, 'dircount':].copy()
x['starpower'] = MinMaxScaler().fit_transform(x['starpower'].astype(float).reshape(-1, 1))
### standardize all other features
y = top_films4['imdbRating']
```

```python
from sklearn.ensemble import RandomForestRegressor 

lr = LinearRegression()
#x = top_films4[['dircount', 'writcount', 'a1count', 'a2count', 'a3count', 'a4count', 'starpower']]
#y = top_films4[['imdbRating']]
lr.fit(x,y)
lr.predict(x.iloc[5])
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.cross_validation import KFold, cross_val_score


cv = KFold(len(y), shuffle=True)

dt = DecisionTreeRegressor()

scores = cross_val_score(dt, x, y, cv=cv)
print(scores)
```

```python
def do_plot(model, m=None):
    for fold, color in zip(cv, ['r', 'g', 'b']):
        X_train = x.iloc[4]
        X_test  = x.iloc[5]
        y_train = y.iloc[4]
        y_test  = y.iloc[5]
        model.fit(x, y)
        y_pred = model.predict(X_train)

        plt.scatter(y_test, y_pred, color=color)
        plt.plot([8.5,9.5],[8.5,9.5])
        plt.text(9,9.5, "R2: "+str(m), fontsize=20, )
        
do_plot(dt, scores.mean().round(2))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

scores = cross_val_score(rf, x, y, cv=cv)
print(scores)
```

```python
from sklearn.model_selection import cross_val_predict

predict = cross_val_predict(rf, x, y, cv=cv)
print(predict)
```

```python
fig, ax = plt.subplots()
ax.scatter(y, predict)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
do_plot(rf, scores.mean().round(2))
```

```python
from sklearn.ensemble import AdaBoostRegressor


ab = AdaBoostRegressor()

scores = cross_val_score(ab, x, y, cv=cv)
scores
```

```python
do_plot(ab, scores.mean().round(2))
```

```python
from sklearn.ensemble import GradientBoostingRegressor


gb = GradientBoostingRegressor()

scores = cross_val_score(gb, x, y, cv=cv)
scores
```
